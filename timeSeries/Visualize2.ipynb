```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/starbucks.csv', index_col='Date', parse_dates=True)
```

```python
df.index
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.plot()
```

```python
df['Close'].plot();
```

```python
df['Volume'].plot()
```

```python
title = "TITLE "
ylabel = 'Y LABEL'
xlabel = 'X LABEL'
```

```python
ax = df['Close'].plot(figsize=(12,5), title=title)
ax.autoscale(axis='x', tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,4))
```

```python
df['Close'].plot(figsize=(12,4), xlim=['2017-01-01','2017-12-31'],ylim=[50,65],ls='--')
```

```python
from matplotlib import dates
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],figsize=(12,5))
ax.set(xlabel=' ')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%a-%B-%d'))

```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'],ylim=[50,60],figsize=(12,5))
ax.set(xlabel=' ')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
